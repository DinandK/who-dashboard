```python
india = pd.read_csv('india_raw.csv', low_memory=False)
india
```

```python
# Cleaning up the data
india.drop(['stn_code', 'agency', 'sampling_date', 'location_monitoring_station'], axis=1, inplace=True) #dropping columns that aren't required
india = india.dropna(subset=['date']) # dropping rows where no date is available
```

```python
types = {
    "Residential": "R",
    "Residential and others": "RO",
    "Residential, Rural and other Areas": "RRO",
    "Industrial Area": "I",
    "Industrial Areas": "I",
    "Industrial": "I",
    "Sensitive Area": "S",
    "Sensitive Areas": "S",
    "Sensitive": "S",
    np.nan: "RRO"
}
india.type = india.type.replace(types)
```

```python
india['date'] = pd.to_datetime(india['date'], errors='coerce')
india.head(5)
```

```python
# no2 status
statewise_no2 = india[['no2','state']].groupby('state', as_index=False).median().sort_values(by='no2')
statewise_no2.head(10)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="amigo")
lat = []
long = []
for row in india['state']:
    loc = geolocator.geocode(row, timeout=10)
    if loc is None:
        lat.append(None)
        long.append(None)
    else:
        latitude = loc.latitude
        longitude = loc.longitude  
    
    lat.append(latitude)
    long.append(longitude)


india['latitude'] = pd.Series(lat)
india['longitude'] = pd.Series(long)
india
```

```python
india.to_csv('india.csv')
```
